```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline

sns.set_theme(style="whitegrid")

df = sns.load_dataset("iris")
print("Dataset loaded successfully!")
```

```python
print("=== DATASET STRUCTURE ===")
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Columns: {df.columns.tolist()}\n")

print("=== FIRST 5 ROWS ===")
display(df.head()) 

print("\n=== DATASET INFO ===")
df.info()

print("\n=== SUMMARY STATISTICS ===")
display(df.describe())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

sns.scatterplot(
    ax=axes[0],
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette="deep",
)
axes[0].set_title("Sepal Length vs. Sepal Width", fontsize=14)
axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Sepal Width (cm)")

sns.histplot(
    ax=axes[1],
    data=df,
    x="petal_length",
    hue="species",
    kde=True,
    multiple="stack",
    palette="deep",
)
axes[1].set_title("Distribution of Petal Length", fontsize=14)
axes[1].set_xlabel("Petal Length (cm)")

sns.boxplot(
    ax=axes[2],
    data=df,
    x="species",
    y="petal_width",
    hue="species", 
    palette="pastel",
    legend=False, 
)
axes[2].set_title("Petal Width Distribution by Species", fontsize=14)
axes[2].set_xlabel("Species")
axes[2].set_ylabel("Petal Width (cm)")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue="species", palette="deep", height=2.5)
plt.suptitle("Comprehensive Pairplot of Iris Features", y=1.02, fontsize=16)
plt.show()
```
